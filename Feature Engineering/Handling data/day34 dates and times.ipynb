# Working with Dates

```python
date['date']= pd.to_datetime(date['date'])
```

```python
date.info()
```

```python
# 1. Extracting year

date['date_year']= date['date'].dt.year

date.head()
```

```python
# 2. Extracting month

date['date_month']= date['date'].dt.month

date.head()
```

```python
# 3. Extracting month_name

date['date_month_name']= date['date'].dt.month_name()

date.head()
```

```python
# 4. Extracting days

date['date_day']= date['date'].dt.day

date.head()
```

```python
# 5. Extracting day of week

date['date_dow']= date['date'].dt.dayofweek

date.head()
```

```python
# 6. Extracting week-name

date['date__dow_name']= date['date'].dt.day_name()

date.head()
```

```python
# 7. is weekend

date['date_is_weekend']= np.where(date['date__dow_name'].isin(['Saturday','Sunday']), 1,0)
date.drop(columns=['product_id','city_id','orders']).head()
date.head()
```

```python
# 8. Extracting week of year

date['date_weekofyear']= date['date'].dt.weekofyear

date.head()
```

```python
# 9. Extracting week-name

date['date_quarter']= date['date'].dt.quarter

date.head()
```

```python
# 7. is weekend

date['date_semester']= np.where(date['date_quarter'].isin([1,2]), 1,2)
date.drop(columns=['product_id','city_id','orders']).head()
date.head()
```

```python
# Extract Time elapsed between dates

import datetime

today = datetime.datetime.today()
today
```

```python
today-date['date']
```

```python
(today-date['date']).dt.days
```

```python
np.round((today-date['date'])/np.timedelta64(1,'M'),0)
```
