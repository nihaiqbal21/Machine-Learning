```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X, y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=80, random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

## OLS

```python
from sklearn.linear_model import LinearRegression
reg= LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X), color='red')
```

## Gradient Descent

```python
# assuming m=78.35, starting value of intercept b=0

y_pred = ((78.35*X)+0).reshape(4)
```

```python
y_pred
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X), color='red', label='OLS')
plt.plot(X,y_pred, color='green', label='b=0')
plt.legend()
plt.show()
```

```python
m = 78.35
b= 0
loss_slope=-2 * np.sum(y- m*X.ravel() - b)
loss_slope
```

```python
# starting with learning rate=0.1
lr=0.1
step_size= loss_slope*lr
step_size
```

```python
#calculating nre intercept
b=b-step_size
b
```

```python
y_pred1 = ((78.35 * X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X), color='red', label='OLS')
plt.plot(X,y_pred1, color='green', label='b={}'.format(b))
plt.plot(X,y_pred, color='lightgreen', label='b=0')
plt.legend()
plt.show()
```

```python
#iteration2
loss_slope= -2 * np.sum(y - m*X.ravel()-b)
print("loss slppe: ", loss_slope)
step_size= loss_slope*lr
print("step size: ", step_size)
b=b-step_size
print("b: ",b)
```

```python
y_pred2 = ((78.35 * X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X), color='red', label='OLS')
plt.plot(X,y_pred2, color='green', label='b={}'.format(b))
plt.plot(X,y_pred1, color='lightblue', label='b={}'.format(b))
plt.plot(X,y_pred, color='lightblue', label='b=0')
plt.legend()
plt.show()
```

```python
#iteration3
loss_slope= -2 * np.sum(y - m*X.ravel()-b)
print("loss slppe: ", loss_slope)
step_size= loss_slope*lr
print("step size: ", step_size)
b=b-step_size
print("b: ",b)
```

```python
y_pred3 = ((78.35 * X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X), color='red', label='OLS')
plt.plot(X,y_pred3, color='green', label='b={}'.format(b))
plt.plot(X,y_pred2, color='lightblue', label='b={}'.format(b))
plt.plot(X,y_pred1, color='lightblue', label='b={}'.format(b))
plt.plot(X,y_pred, color='lightblue', label='b=0')
plt.legend()
plt.show()
```

```python
b = -100
m = 78.35
lr=0.01
epochs=100

for i in range(epochs):
    loss_slope = -2 * np.sum( y-m*X.ravel()-b)
    b=b-(lr* loss_slope)
    
    y_pred = m*X+b
    plt.plot(X,y_pred)
    
plt.scatter(X,y)
```
